### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
b1D = np.arange(start=10, stop=50)
print(b1D)
b1D = np.flip(b1D)
print(b1D)
print("Все четные элементы:")
for x in b1D:
    if not x & 1:
        print(x, end=" ")

print("")
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
e2D_1=np.random.rand(6,4)
print(e2D_1) 
print()
e2D_2=np.random.rand(4,3)
print(e2D_2) 
print()
print(np.matmul(e2D_1,e2D_2))
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
np.random.seed(0)  # Для воспроизводимости результатов
array = np.random.rand(7, 7)

# Вычисляем среднее и стандартное отклонение
mean = np.mean(array)
std = np.std(array)

# Нормализуем массив
normalized_array = (array - mean) / std

print("Исходный массив:\n", array)
print("\nСреднее значение:", mean)
print("Стандартное отклонение:", std)
print("\nНормализованный массив:\n", normalized_array)
```

```python
print(tips_df.groupby('day', observed=False)['tip_percent'].mean().sort_values())
```

```python
print(tips_df.groupby('time', observed=False)['tip_percent'].mean(),)
```
